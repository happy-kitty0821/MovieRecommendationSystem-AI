```python
#setting the datasets into movies and credits dataframe
movies = pandas.read_csv("datasets/tmdb_5000_movies.csv")
credits = pandas.read_csv("datasets/tmdb_5000_credits.csv") 
```

```python
#merging the 'movies' DataFrame with the 'credits' DataFrame on the 'title' column.
#this will combines rows from both DataFrames where the 'title' column matches in both.
#the resulting DataFrame will contain columns from both DataFrames for matching movie titles.
movies = movies.merge(credits, on='title')
```

```python
#selecting a subset of columns from the 'movies' DataFrame.
#this will keeps only the columns 'movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', and 'crew'.
#the resulting DataFrame will only contain these columns, making it simpler and more focused on the required data.
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
def convert(text):
    #initialize an empty list to store names
    L = []
    #safely evaluate the string 'text' to a Python object (e.g., list of dictionaries)
    for i in ast.literal_eval(text):
        #append the value of the 'name' key from each dictionary to the list
        L.append(i['name']) 
    #return the list of names
    return L 
```

```python
movies.dropna(inplace=True)#this will drop all the rows containing nul or nah values
```

```python
#apply the 'convert' function to the 'genres' column to transform each entry
#from a string representation of a list of dictionaries to a list of genre names.
movies['genres'] = movies['genres'].apply(convert)
#display the first five rows of the modified DataFrame to verify the transformation.
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    # Initialize an empty list to store names
    L = []
    # Initialize a counter to limit to the first three items
    counter = 0
    # Safely evaluate the string 'text' to a Python object (e.g., list of dictionaries)
    for i in ast.literal_eval(text):
        # Append the value of the 'name' key from each dictionary to the list if counter is less than 3
        if counter < 3:
            L.append(i['name'])
        #increment the counter
        counter += 1
    #return the list of names
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    # Initialize an empty list to store modified strings
    L1 = []
    # Iterate over each string in the input list L
    for i in L:
        # Remove all spaces from the string and append the modified string to L1
        L1.append(i.replace(" ", ""))
    # Return the list of modified strings
    return L1
```

```python
#apply the 'collapse' function to the 'cast' column to remove spaces from each name in the list
movies['cast'] = movies['cast'].apply(collapse)
#apply the 'collapse' function to the 'crew' column to remove spaces from each name in the list
movies['crew'] = movies['crew'].apply(collapse)
#apply the 'collapse' function to the 'genres' column to remove spaces from each genre name in the list
movies['genres'] = movies['genres'].apply(collapse)
#apply the 'collapse' function to the 'keywords' column to remove spaces from each keyword in the list
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()

### **Count Vectorizor**

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
cv
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

### **finding cosine similarity**

```python
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
"""this function, recommend, takes a movie title as input and recommends five similar movies based on some precomputed similarity scores.
"""
def recommend(movie):
    try:
        index = new[new['title'] == movie].index[0]
        distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
        for i in distances[1:6]:
            print(new.iloc[i[0]]['title'])
    except IndexError:
        print(f"Could not find movie '{movie}' in the dataset.")
```

### **demo usage on calling of the recommend function**

```python
recommend("Titanic")
```
